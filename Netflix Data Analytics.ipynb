### Q11) How can we sort the dataset by year

```python
df.head(3)
```

```python
# Fetch the Year from thr "Release_Date" Column
df['Year'] = df['Release_Date'].dt.year
```

```python
# Now convert the "Year" column float To Int
df['Year'] = df['Year'].astype(int)
```

```python
# sorting the dataset DESC to ASC
df.sort_values(by ='Year',ascending=False) 
```

```python
# sorting the dataset ASC to DESC

df.sort_values(by = 'Year',ascending=True)
```
